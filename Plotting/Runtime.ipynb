```python
import csv
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as md

hms = md.DateFormatter('%H:%M:%S')

files = ["results-before.csv", "results-after.csv", "results-doubling.csv"]
for i in files:
    x,y = [],[]
    csv_reader = csv.reader(open(i))
    for line in csv_reader:
        x.append(int(line[0]))
        y.append(dt.datetime.strptime(line[1],'%H:%M:%S'))

    fig = plt.figure()
    ax = fig.add_subplot()
    ax.plot(x,y,'o-')
    ax.yaxis.set_major_formatter(hms)

    plt.xlabel('n of images')
    plt.ylabel('runtime')
    plt.tight_layout()

    plt.savefig("total-runtime" + str(i) + ".png")
```
