```python
import csv
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib.dates as md

files = ["results-before.csv", "results-after.csv"]
count = 0

for i in files:
    count += 1
    x, y = [], []
    csv_reader = csv.reader(open(i))
    for line in csv_reader:
        n = int(line[0])
        time = dt.datetime.strptime(line[1], '%H:%M:%S')
        totalseconds = (time.hour * 60 + time.minute) * 60 + time.second
        timeperImage = totalseconds/n
        x.append(n)
        y.append(timeperImage)

    plt.plot(x,y,'o-', label="experiment " + str(count))
    plt.legend

plt.xlabel('n of images')
plt.ylabel('runtime per image in s')
plt.savefig("runtime-per-image-combined.png")
```
